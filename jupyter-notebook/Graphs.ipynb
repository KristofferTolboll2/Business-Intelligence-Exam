```python
from dataclasses import dataclass

@dataclass(order=True)
class User:
    """Generated class for users"""
    _id: int
    mental_health_level: int
    membership_days: int = 0

```

```python

users = []

def generate_user_data(user_amount = 1000):
    values = range(0, 100, 10)
    for userId in range(user_amount):
        user = User(userId, random.choice(values), random.choice(values))
        print(user)
        users.append(user)
        
def clear_users():
    users = []
        
        
```

```python
generate_user_data(10)
```

```python
clear_users()
```

```python
def generate_graph(users):
    fig = plt.figure()
    ax = plt.axes(projection="3d")
    
    x_vals = list(map(lambda user: user.mental_health_level, users))
    y_vals = list(map(lambda user: user.membership_days, users))
    z_vals = list(map(lambda user: user._id, users))
    
    x_vals.sort()
    y_vals.sort()
    z_vals.sort()
    
    
    num_bars = len(users)
    x_pos = x_vals
    y_pos = y_vals
    z_pos = z_vals
    x_size = np.ones(num_bars)
    y_size = np.ones(num_bars)
    z_size = random.sample(range(num_bars), num_bars)

    ax.bar3d(x_pos, y_pos, z_pos, x_size, y_size, z_size, color='aqua')
    plt.show()
```

```python
generate_graph(users)
```
